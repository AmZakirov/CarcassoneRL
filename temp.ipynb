Winning rate calculation:

```python
def winning_rate(winners, episodes):
    new_winners : np.array = deepcopy(winners[episodes:])
    return  round(Counter(new_winners)[0]/ len(new_winners), 3)

Counter(winners)[0] / len(winners)
```

```python
def find_max_value(dictionary_list, key):
    max_dict = max(dictionary_list, key=lambda x: x[key])
    return max_dict

# Пример использования
dictionary_list = [
    {'a': 10, 'b': 25},
    {'a': 5, 'b': 15},
    {'a': 8, 'b': 25}
]

max_dict = find_max_value(dictionary_list, 'b')
print(max_dict)
```
